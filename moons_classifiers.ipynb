```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
n_samples = 1500
centers = [(0, 0), (5, 5)]
X, y = make_moons(n_samples=n_samples,noise=0.1)

plt.figure(figsize=(7,7))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Data")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0,train_size=0.8)
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis, LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
qda = QuadraticDiscriminantAnalysis(priors=None)
lda = LinearDiscriminantAnalysis(priors=None)
```

```python
gnb_pred = gnb.fit(Xtrain, ytrain).predict(Xtest)
qda_pred = qda.fit(Xtrain, ytrain).predict(Xtest)
lda_pred = lda.fit(Xtrain, ytrain).predict(Xtest)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(ytest, gnb_pred))
print()
print(confusion_matrix(ytest, qda_pred))
print()
print(confusion_matrix(ytest, lda_pred))
```

```python
print(classification_report(ytest, gnb_pred))
print()
print(classification_report(ytest, qda_pred))
print()
print(classification_report(ytest, lda_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn_pred = knn.fit(Xtrain, ytrain).predict(Xtest)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(ytest, knn_pred))
print()
print(classification_report(ytest, knn_pred))
```
