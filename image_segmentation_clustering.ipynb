```python
from skimage.io import imread

I = imread('/content/imagetest.jpg')
```

```python
I.shape
```

```python
import numpy as np

X = np.reshape(I,(I.shape[0]*I.shape[1],I.shape[2]))
X.shape
```

```python
from sklearn.cluster import KMeans

labels = KMeans(n_clusters=3).fit_predict(X)

labels.shape
```

```python
L = np.reshape(labels,(I.shape[0],I.shape[1]))

```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(5,5))
ax1 = fig.add_subplot(121)
ax1 = plt.imshow(I)

ax2 = fig.add_subplot(122)
ax2 = plt.imshow(L)
```

```python
from sklearn.cluster import AgglomerativeClustering

labels1 = AgglomerativeClustering(n_clusters=3).fit_predict(X)
L1 = np.reshape(labels1,(I.shape[0],I.shape[1]))
fig = plt.figure(figsize=(5,5))
ax1 = fig.add_subplot(121)
ax1 = plt.imshow(I)

ax2 = fig.add_subplot(122)
ax2 = plt.imshow(L1)
```
